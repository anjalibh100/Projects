# Scraping Data from Wikipedia (Real time data analysis) and writing it to a csv file

# Access Website to get "Top publicly traded Indian companies by revenue" 

**1. First import all the necessary python libraries. This includes BeautifulSoap and csv libraries**

```python
#Import the necessary libraries
from bs4 import BeautifulSoup
import requests
```

**2. Specify the URL. This is required to navigate the website (In the URL) and access the data.**

```python
url = "https://companiesmarketcap.com/india/largest-indian-companies-by-revenue/"
#url = "https://en.wikipedia.org/wiki/List_of_largest_companies_in_India"
page=requests.get(url)
#soup = BeautifulSoup(page.content,'lxml')
soup=BeautifulSoup(page.content,"html.parser") ##For formatted output
```

**3. View the page content where the URL points**

```python
print(soup)
```

```python
#Find out the number of tables 
how_many_tables=soup.find_all('table')
Table_numbers=len(how_many_tables)
print(Table_numbers)
```

**The result shows only 1 table exists in the page content**

```python
#comp_titles=soup.find('table',class_='default-table')
table = soup.find_all('table')[0]
print(table.prettify())
```

****Find all tags with 'td' ****

```python
Table_content = soup.find('table',class_='default-table')
#print(Table_content)
col_data = Table_content.find_all('td')
print(col_data)
```

```python
#Print all the Company Names ( a tags)
for row in col_data:
    tags_a = row.find_all('a')
    for med in tags_a:
        #print(med.text)
        print(med)
        
```

```python
soup2 = soup.find('table',class_='default-table')
soup2 = soup.prettify()
print(soup2)
```

```python
Table_content = soup.find('table',class_='default-table')
col_data = Table_content.find_all('td')
individual_row_data = [data.text.strip() for data in col_data]
#print(individual_row_data)
for i in (0,100):
    individual_row_data.remove('Close Ad X')  #Cleaning the data
print(individual_row_data)
```

```python
print(len(individual_row_data))
```

```python
#Finally getting the data and saving it to CSV
import csv

Data_format = []


counter = 0
for z in range (0,800,8):
    v = z
    w = v + 8
    Data_format = individual_row_data[v:w]
    Data_format.pop(0)
    Company_Name = ' '.join(individual_row_data[v+2].strip().replace('\n', '').split())
    print(Data_format)
    file = open('out_scrape2.csv', 'a', newline ='')
    with file:
    # identifying header  
        header = ['Rank', 'Name', 'Revenue','Price']
        writer = csv.DictWriter(file, fieldnames = header)
        writer.writerow({'Rank' : individual_row_data[v+1],'Name': Company_Name,'Revenue': individual_row_data[v+3],'Price': individual_row_data[v+4]})



```
