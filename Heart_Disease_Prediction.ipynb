```python
correlation_matrix = med_dataset.corr()
print(correlation_matrix)
sns.heatmap(correlation_matrix,annot=True,cmap = "coolwarm")
```

Inferences on this data indicates -

Age and Max HR have a negative correlation of -0.4022, indicating that as age increases, the maximum heart rate tends to decrease.
Cholesterol and Sex have a negative correlation of -0.2016, suggesting that there is a moderate negative relationship between cholesterol levels and gender.
Exercise angina and Chest pain type have a strong positive correlation of 0.3532, indicating that as the type of chest pain increases, the likelihood of exercise-induced angina also increases.
ST depression and Slope of ST have a strong positive correlation of 0.6097, implying that as the ST segment depression increases, the slope of the ST segment also tends to increase.
Number of vessels fluro and Age have a positive correlation of 0.3561, suggesting that the number of vessels appearing fluoroscopy increases with age.
Thallium and Sex have a strong positive correlation of 0.3910, indicating a significant relationship between thallium levels and gender.
BP and Age show a positive correlation of 0.2731, suggesting that blood pressure tends to increase with age.
Max HR and Exercise angina have a negative correlation of -0.3807, indicating that as the maximum heart rate decreases, the likelihood of exercise-induced angina increases.
ST depression and Max HR have a negative correlation of -0.3490, suggesting that as the maximum heart rate decreases, the ST segment depression tends to increase.
Cholesterol and Thallium show a weak positive correlation of 0.0288, suggesting a minor relationship between cholesterol levels and thallium.


```python
num_cols=med_dataset.select_dtypes(include=float)

plt.figure(figsize=(28,10))
for i,j in zip(range(1, 10),num_cols):
    plt.subplot(3, 3, i)
    plt.subplots_adjust(right=0.9,top=1.8)
    sns.boxplot(data=med_dataset, x=j)
    sns.set_theme()
    plt.title('Boxplot of {}'.format(j))
```
